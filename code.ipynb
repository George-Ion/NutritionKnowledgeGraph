```python
from rdflib import Graph, Namespace, URIRef, Literal
from rdflib.namespace import RDF, RDFS, OWL, XSD

g = Graph()

# Namespaces
SDO = Namespace("https://schema.org/")
g.bind("schema", SDO)

# Classes
g.add((SDO.Recipe, RDF.type, OWL.Class))
g.add((SDO.Review, RDF.type, OWL.Class))
g.add((SDO.Nutrition, RDF.type, OWL.Class))
g.add((SDO.Restaurant, RDF.type, OWL.Class))

# Recipe properties
g.add((SDO.image, RDF.type, OWL.DatatypeProperty))
g.add((SDO.image, RDFS.domain, SDO.Recipe))
g.add((SDO.image, RDFS.range, XSD.anyURI))

g.add((SDO.keywords, RDF.type, OWL.DatatypeProperty))
g.add((SDO.keywords, RDFS.domain, SDO.Recipe))
g.add((SDO.keywords, RDFS.range, XSD.string))

g.add((SDO.datePublished, RDF.type, OWL.DatatypeProperty))
g.add((SDO.datePublished, RDFS.domain, SDO.Recipe))
g.add((SDO.datePublished, RDFS.range, XSD.dateTime))

g.add((SDO.cookTime, RDF.type, OWL.DatatypeProperty))
g.add((SDO.cookTime, RDFS.domain, SDO.Recipe))
g.add((SDO.cookTime, RDFS.range, XSD.string))

g.add((SDO.hasReview, RDF.type, OWL.ObjectProperty))
g.add((SDO.hasReview, RDFS.domain, SDO.Recipe))
g.add((SDO.hasReview, RDFS.range, SDO.Review))

g.add((SDO.hasNutrition, RDF.type, OWL.ObjectProperty))
g.add((SDO.hasNutrition, RDFS.domain, SDO.Recipe))
g.add((SDO.hasNutrition, RDFS.range, SDO.Nutrition))

g.add((SDO.hasCuisine, RDF.type, OWL.ObjectProperty))
g.add((SDO.hasCuisine, RDFS.domain, SDO.Recipe))
g.add((SDO.hasCuisine, RDFS.range, SDO.Restaurant))

# Nutrition properties
g.add((SDO.calories, RDF.type, OWL.DatatypeProperty))
g.add((SDO.calories, RDFS.domain, SDO.Nutrition))
g.add((SDO.calories, RDFS.range, XSD.decimal))

g.add((SDO.sugarContent, RDF.type, OWL.DatatypeProperty))
g.add((SDO.sugarContent, RDFS.domain, SDO.Nutrition))
g.add((SDO.sugarContent, RDFS.range, XSD.decimal))

g.add((SDO.proteinContent, RDF.type, OWL.DatatypeProperty))
g.add((SDO.proteinContent, RDFS.domain, SDO.Nutrition))
g.add((SDO.proteinContent, RDFS.range, XSD.decimal))

g.add((SDO.fiberContent, RDF.type, OWL.DatatypeProperty))
g.add((SDO.fiberContent, RDFS.domain, SDO.Nutrition))
g.add((SDO.fiberContent, RDFS.range, XSD.decimal))

# Restaurant properties
g.add((SDO.aggregateRating, RDF.type, OWL.DatatypeProperty))
g.add((SDO.aggregateRating, RDFS.domain, SDO.Restaurant))
g.add((SDO.aggregateRating, RDFS.range, XSD.decimal))

g.add((SDO.priceRange, RDF.type, OWL.DatatypeProperty))
g.add((SDO.priceRange, RDFS.domain, SDO.Restaurant))
g.add((SDO.priceRange, RDFS.range, XSD.decimal))

g.add((SDO.city, RDF.type, OWL.DatatypeProperty))
g.add((SDO.city, RDFS.domain, SDO.Restaurant))
g.add((SDO.city, RDFS.range, XSD.string))

g.add((SDO.country, RDF.type, OWL.DatatypeProperty))
g.add((SDO.country, RDFS.domain, SDO.Restaurant))
g.add((SDO.country, RDFS.range, XSD.string))

g.add((SDO.hasDeliveryService, RDF.type, OWL.DatatypeProperty))
g.add((SDO.hasDeliveryService, RDFS.domain, SDO.Restaurant))
g.add((SDO.hasDeliveryService, RDFS.range, XSD.boolean))

print(g.serialize(format="turtle"))

```

```python
fod = Namespace("http://kg-course.io/food-nutrition/")
g.bind("fod", fod)
```

```python
import pandas as pd

nutrion_df = pd.read_csv(
    "C:/Users/georg/Desktop/Master/Period 4/Building and mining knowledge graphs/Assignment 1/data/Nutrition.csv",
    sep=";",
    encoding="latin1"
)
nutrion_df.head()
```

```python
for _,row in nutrion_df.iterrows():
     
    # Transform the names into URIs
    clean_name = row["Name"].strip().replace('"', '').replace("'", "").replace(" ", "_").replace(",", "")   
    # Using "http://kg-course.io/food-nutrition/" as the namespace for the food items 
    # # We create a URI for each food item and add it to the graph as an instance of the Nutrition class
    row["Name"] = fod[clean_name]
    # Add the nutrition information to the graph as RDF triples, calories, sugar content, protein content, and fiber content are added
    # as properties from schema.org
    g.add((URIRef(row["Name"]), RDF.type, SDO.Nutrition))
    g.add((URIRef(row["Name"]), SDO.calories, Literal(row["Calories"], datatype=XSD.decimal)))
    g.add((URIRef(row["Name"]), SDO.sugarContent, Literal(row["SugarContent"], datatype=XSD.decimal)))
    g.add((URIRef(row["Name"]), SDO.proteinContent, Literal(row["ProteinContent"], datatype=XSD.decimal)))
    g.add((URIRef(row["Name"]), SDO.fiberContent, Literal(row["FiberContent"], datatype=XSD.decimal)))

    if row.name > 1000:
        break

print(g.serialize(format='ttl'))
```

```python
import pandas as pd

recipes_df = pd.read_csv(
    "C:/Users/georg/Desktop/Master/Period 4/Building and mining knowledge graphs/Assignment 1/data/Recipes.csv",
    sep=";",
    encoding="latin1",
    #Trouble with multi-line cells, so we use the Python engine and specify the quote character to handle the c("") blocks properly
    engine='python',        
    quotechar='"',          
    # default is doublequote=True, but we specify it explicitly to ensure that the "" inside the c("") blocks are handled correctly
    doublequote=True,       
    # In case of a bad line we skip it
    on_bad_lines='skip'     
)

recipes_df.head()
```

```python
for _,row in recipes_df.iterrows():
     
    # Transform the names into URIs
    clean_name = row["Name"].strip().replace('"', '').replace("'", "").replace(" ", "_").replace(",", "")
    recipe_uri = fod[clean_name]
    
    # Add BEFORE converting to URI
    original_name = row["Name"]
    
    # Add the recipe to the graph as RDF triples
    g.add((recipe_uri, RDF.type, SDO.Recipe))
    g.add((recipe_uri, SDO.name, Literal(original_name, datatype=XSD.string)))
    g.add((recipe_uri, SDO.image, Literal(row["Images"], datatype=XSD.anyURI)))
    g.add((recipe_uri, SDO.keywords, Literal(row["Keywords"], datatype=XSD.string)))
    g.add((recipe_uri, SDO.datePublished, Literal(row["DatePublished"], datatype=XSD.dateTime)))
    g.add((recipe_uri, SDO.cookTime, Literal(row["CookTime"], datatype=XSD.string)))

    if row.name > 1000:
        break

g.serialize("recipes_with_nutrition_links2.ttl", format="turtle")
print(g.serialize(format='ttl'))
```

```python
import pandas as pd

restaurants_df = pd.read_csv(
    "C:/Users/georg/Desktop/Master/Period 4/Building and mining knowledge graphs/Assignment 1/data/Restaurants.csv",
    sep=";",
    encoding="latin1", 
)

restaurants_df.head()
```

```python
for _, row in restaurants_df.iterrows():
    clean_name = row["Restaurant Name"].strip().replace('"', '').replace("'", "").replace(" ", "_").replace(",", "").replace("{" , "").replace("}", "")
    restaurant_uri = fod[clean_name]
    
    # Add restaurant as instance
    g.add((restaurant_uri, RDF.type, SDO.Restaurant))
    g.add((restaurant_uri, SDO.aggregateRating, Literal(row["Aggregate rating"], datatype=XSD.decimal)))
    g.add((restaurant_uri, SDO.priceRange, Literal(row["Price range"], datatype=XSD.decimal)))
    g.add((restaurant_uri, SDO.city, Literal(row["City"], datatype=XSD.string)))
    g.add((restaurant_uri, SDO.country, Literal(row["Country"], datatype=XSD.string)))
    g.add((restaurant_uri, SDO.hasDeliveryService, Literal(row["Has Online delivery"], datatype=XSD.boolean)))
    g.add((restaurant_uri, SDO.cuisineType, Literal(row["Cuisines"], datatype=XSD.string)))
    
    if row.name > 1000:
        break

g.serialize("restaurants.ttl", format="turtle")
print(g.serialize(format='ttl'))
```

```python
# ============================================================================
# SPARQL QUERIES - Optimized for current data structure
# ============================================================================
# These queries work WITHOUT requiring schema:name to be present
# They extract names from URIs when needed

from rdflib.plugins.sparql import prepareQuery

print("\n" + "="*80)
print("WORKING QUERIES (Tasks 1-2 Data) - OPTIMIZED")
print("="*80)

# ============================================================================
# QUERY 1: Find recipes by keywords - FIXED (works without schema:name)
# ============================================================================
query_recipes_by_keyword = """
PREFIX schema: <https://schema.org/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT ?recipe ?keywords ?cookTime
WHERE {
  ?recipe rdf:type schema:Recipe .
  ?recipe schema:keywords ?keywords .
  ?recipe schema:cookTime ?cookTime .
  FILTER(CONTAINS(LCASE(?keywords), "pie"))
}
LIMIT 20
"""

print("\n[QUERY 1] Find all PIE recipes:")
print("-" * 80)
results = g.query(query_recipes_by_keyword)
print(f"Found {len(results)} pie recipes:\n")
for row in results:
    recipe_name = str(row.recipe).split('/')[-1]  # Extract name from URI
    print(f"   {recipe_name}")
    print(f"    Keywords: {row.keywords}")
    print(f"    Cook Time: {row.cookTime}\n")

# ============================================================================
# QUERY 2: Find recipes by publication year - FIXED
# ============================================================================
query_recipes_by_year = """
PREFIX schema: <https://schema.org/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?recipe ?published
WHERE {
  ?recipe rdf:type schema:Recipe .
  ?recipe schema:datePublished ?published .
  FILTER(YEAR(?published) > 2000)
}
ORDER BY DESC(?published)
LIMIT 30
"""

print("\n[QUERY 2] Find recipes published after 2000:")
print("-" * 80)
results = g.query(query_recipes_by_year)
print(f"Found {len(results)} recipes published after 2000:\n")
for row in results:
    recipe_name = str(row.recipe).split('/')[-1]
    year = row.published.year
    print(f"   {recipe_name} ({year})")

# ============================================================================
# QUERY 3: Find nutrition data by specific nutrient
# ============================================================================
query_nutrition_search = """
PREFIX schema: <https://schema.org/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT ?nutrition ?calories ?protein ?fiber ?sugar
WHERE {
  ?nutrition rdf:type schema:Nutrition .
  ?nutrition schema:calories ?calories .
  ?nutrition schema:proteinContent ?protein .
  ?nutrition schema:fiberContent ?fiber .
  ?nutrition schema:sugarContent ?sugar .
  FILTER(?protein > 15)
}
ORDER BY DESC(?protein)
LIMIT 20
"""

print("\n[QUERY 3] Find high-protein foods (protein > 15g):")
print("-" * 80)
results = g.query(query_nutrition_search)
print(f"Found {len(results)} high-protein foods:\n")
for row in results:
    nutrition_name = str(row.nutrition).split('/')[-1]
    print(f"   {nutrition_name}")
    print(f"    Protein: {row.protein}g | Calories: {row.calories} | Fiber: {row.fiber}g")

# ============================================================================
# QUERY 4: Find restaurants in a specific city
# ============================================================================
query_restaurants_by_city = """
PREFIX schema: <https://schema.org/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT ?restaurant ?city ?rating ?cuisine ?delivery
WHERE {
  ?restaurant rdf:type schema:Restaurant .
  ?restaurant schema:city ?city .
  ?restaurant schema:aggregateRating ?rating .
  ?restaurant schema:cuisineType ?cuisine .
  ?restaurant schema:hasDeliveryService ?delivery .
  FILTER(CONTAINS(?city, "Delhi"))
}
ORDER BY DESC(?rating)
LIMIT 20
"""

print("\n[QUERY 4] Find restaurants in Delhi with ratings:")
print("-" * 80)
results = g.query(query_restaurants_by_city)
print(f"Found {len(results)} restaurants in Delhi:\n")
for row in results:
    restaurant_name = str(row.restaurant).split('/')[-1]
    delivery_status = " Yes" if row.delivery else " No"
    print(f"   {restaurant_name}")
    print(f"    City: {row.city} | Rating: {row.rating} | Cuisine: {row.cuisine}")
    print(f"    Online Delivery: {delivery_status}\n")

# ============================================================================
# QUERY 5: Find restaurants with online delivery - FIXED
# ============================================================================
query_delivery_restaurants = """
PREFIX schema: <https://schema.org/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT ?restaurant ?city ?cuisine ?priceRange
WHERE {
  ?restaurant rdf:type schema:Restaurant .
  ?restaurant schema:city ?city .
  ?restaurant schema:cuisineType ?cuisine .
  ?restaurant schema:priceRange ?priceRange .
  ?restaurant schema:hasDeliveryService ?delivery .
  FILTER(?delivery = true || ?delivery = 1 || ?delivery = "true"^^xsd:string)
}
ORDER BY ?city
LIMIT 30
"""

print("\n[QUERY 5] Find restaurants with ONLINE DELIVERY:")
print("-" * 80)
results = g.query(query_delivery_restaurants)
print(f"Found {len(results)} restaurants with online delivery:\n")
for row in results:
    restaurant_name = str(row.restaurant).split('/')[-1]
    print(f"   {restaurant_name} ({row.city})")
    print(f"    Cuisine: {row.cuisine} | Price: {row.priceRange}\n")

# ============================================================================
# QUERY 6: Find restaurants by cuisine type
# ============================================================================
query_restaurants_by_cuisine = """
PREFIX schema: <https://schema.org/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>

SELECT ?restaurant ?city ?cuisine ?rating
WHERE {
  ?restaurant rdf:type schema:Restaurant .
  ?restaurant schema:city ?city .
  ?restaurant schema:cuisineType ?cuisine .
  ?restaurant schema:aggregateRating ?rating .
  FILTER(CONTAINS(LCASE(?cuisine), "chinese"))
}
ORDER BY DESC(?rating)
LIMIT 20
"""

print("\n[QUERY 6] Find CHINESE restaurants by rating:")
print("-" * 80)
results = g.query(query_restaurants_by_cuisine)
print(f"Found {len(results)} Chinese restaurants:\n")
for row in results:
    restaurant_name = str(row.restaurant).split('/')[-1]
    print(f"   {restaurant_name}")
    print(f"    Location: {row.city} | Rating: {row.rating}\n")
```
